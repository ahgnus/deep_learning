```python
# 이상치(outlier) 확인
# z-score 이용 
def outlier(df, col, z):
    return df[abs(df[col] - np.mean(df[col]))/np.std(df[col])>z].index
print("OK")
```

```python
#북미에서만 특출하게 많이 팔린 이상치가 있는지 점검 z값 1.5로 설정

outlier(vgsales, 'NA_Sales', 1.5)
```

```python
#위치메소드로 나타내줘야됨
vgsales.loc[outlier(vgsales, 'NA_Sales', 1.5)]
```

```python
#521 개나 검색됨 z값을 더 크게 잡아봄  (521/16598) z = 2
vgsales.loc[outlier(vgsales, 'NA_Sales', 2)]
```

```python
#361 개나 검색됨 z값을 더 크게 잡아봄  (361/16598) z = 3
vgsales.loc[outlier(vgsales, 'NA_Sales', 3)]

#174개
```

```python
# outlier 함수의 부등호 값을 반대로 하여 이상치가 아닌값을 나타낼수있음 
def not_outlier(df, col, z):
    return df[abs(df[col] - np.mean(df[col]))/np.std(df[col])<=z].index
print("OK")
```

```python
vgsales.loc[not_outlier(vgsales, 'NA_Sales', 1.5)]
```
